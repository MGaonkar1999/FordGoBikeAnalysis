## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1 : 
Q1. How long does the average trip take?

````python
# Chart - 1 visualization code
# Convert 'start_time' to datetime
fb['start_time'] = pd.to_datetime(fb['start_time'])

# Extract hour
fb['start_hour'] = fb['start_time'].dt.hour

# Calculate average duration by hour
avg_duration_by_hour = fb.groupby('start_hour')['duration_sec'].mean().reset_index()
avg_duration_by_hour['duration_min'] = avg_duration_by_hour['duration_sec'] / 60

# Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_duration_by_hour, x='start_hour', y='duration_min', marker='o', color='teal')

# Add labels on each point
for i in range(len(avg_duration_by_hour)):
    plt.text(
        avg_duration_by_hour['start_hour'][i],
        avg_duration_by_hour['duration_min'][i] + 0.2,  # slightly above the point
        f"{avg_duration_by_hour['duration_min'][i]:.1f}",
        ha='center', fontsize=9
    )

plt.title('Average Trip Duration by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Average Duration (minutes)')
plt.xticks(range(0, 24))
plt.grid(True)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

This Line plot chart shows how trip duration trends over the course of the day, easily spot peaks through Trend visbility. It gives compact yet informative insights.

##### 2. What is/are the insight(s) found from the chart?

Unusually Long Trips in Early Morning (2 AM – 4 AM):Average trip durations peak at 46.8 minutes around 3 AM, which is significantly higher than any other hour of the day.

Commuter Hours Have Shorter Trips (7 AM – 9 AM and 17 PM – 19 PM):Average trip durations are lowest during morning and evening rush hours, suggesting commuter behavior — likely short, predictable routes.

Gradual Rise in Midday:Between 10 AM – 15 PM, there's a moderate increase in duration, possibly from tourists or casual riders taking longer trips.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, It helps in optimizing fleet distribution(Knowing that short trips dominate rush hours, we can ensure more bikes are available at popular commuter hubs during those times.)

In terms of Negative insights, we can see Risk of Operational efficiency(If these long trips (e.g., 3 AM) are also fewer in number, they may skew maintenance scheduling, delaying servicing where needed more urgently).

#### Chart - 2:
Is the trip duration affected by weather?(months/seasons)

````python
# Chart - 2 visualization code
# Extract month
fb['month'] = fb['start_time'].dt.month

# Define function to map seasons
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Create season column
fb['season'] = fb['month'].apply(get_season)

# Calculate average duration by season
avg_duration_by_season = fb.groupby('season')['duration_sec'].mean().reset_index()
avg_duration_by_season['duration_min'] = avg_duration_by_season['duration_sec'] / 60

# Sort seasons
season_order = ['Winter', 'Spring', 'Summer', 'Fall']
avg_duration_by_season['season'] = pd.Categorical(avg_duration_by_season['season'], categories=season_order, ordered=True)
avg_duration_by_season = avg_duration_by_season.sort_values('season')

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(data=avg_duration_by_season, x='season', y='duration_min', palette='Spectral')

# Add labels on bars
for i, row in avg_duration_by_season.iterrows():
    ax.text(i, row['duration_min'] + 0.2, f"{row['duration_min']:.1f}", ha='center', fontsize=10)

plt.title('Average Trip Duration by Season')
plt.xlabel('Season')
plt.ylabel('Average Duration (minutes)')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

This bar chart was chosen as it helps in visually compare durations b/w each seasons & also helps to clearly communicate where trip durations peak or drop.

##### 2. What is/are the insight(s) found from the chart?

According to chart , Winter has highest Avg. trip duration.

3.Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes, it will create a +ve impact as it helps in Marketing campaign & Resource Allocation.

In terms of Negative insights, we can see that in Spring, Summer, and Fall season there is zero average duration due to missing data which may also cause misallocation of resources based on incorrect assumptions (e.g., focusing only on winter).

#### Chart - 8 
What are the most frequent start and end station combinations?

````python
# Chart - 8 visualization code
# Create route column if not already present
fb['route'] = fb['start_station_name'] + " → " + fb['end_station_name']

# Identify top 10 start and end stations
top_starts = fb['start_station_name'].value_counts().nlargest(10).index
top_ends = fb['end_station_name'].value_counts().nlargest(10).index

# Filter data to include only those routes
fb = fb[fb['start_station_name'].isin(top_starts) & fb['end_station_name'].isin(top_ends)]

# Create pivot table
route_matrix = fb.pivot_table(index='start_station_name',
                                       columns='end_station_name',
                                       values='route',
                                       aggfunc='count',
                                       fill_value=0)

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(route_matrix, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5, cbar_kws={'label': 'Trip Count'})
plt.title('Heatmap of Top 10 Start → End Station Combinations')
plt.xlabel('End Station')
plt.ylabel('Start Station')
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Heatmap - Helps to visualize the relationship. Through Color gradients we can quickly highlight which station pairs are the most frequently traveled.

#### Chart - 9: 
Is there a correlation between age and trip duration?

````python
# Chart - 9 visualization code
# Filter out unrealistic ages
fb = fb[(fb['age'] >= 10) & (fb['age'] <= 90)]

# Convert trip duration to minutes
fb['trip_duration_min'] = fb['duration_sec'] / 60

# Plot scatter plot with regression line
plt.figure(figsize=(8, 6))
sns.regplot(data=fb, x='age', y='trip_duration_min',
            scatter_kws={'alpha': 0.3}, line_kws={'color': 'red'})
plt.title('Correlation Between Age and Trip Duration')
plt.xlabel('Age')
plt.ylabel('Trip Duration (minutes)')
plt.tight_layout()
plt.show()

# Calculate and display the correlation coefficient
correlation = fb[['age', 'trip_duration_min']].corr().iloc[0, 1]
print(f"Correlation coefficient: {correlation:.4f}")
````

##### 1. Why did you pick the specific chart?

Scatter plot- Best way to visualize relationship b/w 2 continous variables. The red line helps in quickly spotting the overall direction of the relationship.

##### 2. What is/are the insight(s) found from the chart?

No strong correlation:The points are widely scattered, and the red trend line is almost flat and slightly declining, meaning age and trip duration are not strongly related. Younger users dominate the trip counts(25-40years).Older users tend to have shorter and more consistent trip durations(>60yr). On the basis of Outliers,some trips have extraordinary durations.

### Chart 10 :
How does average trip duration vary by day of the week and Bike Share participation?

````python
# Chart - 9 visualization code

fb['day_of_week'] = fb['start_time'].dt.day_name()

# Define the order of days
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
fb['day_of_week'] = pd.Categorical(fb['day_of_week'], categories=day_order, ordered=True)

# Group by day of week, Bike Share status, and gender, then calculate average trip duration
avg_duration_gender = fb.groupby(['day_of_week', 'bike_share_for_all_trip', 'member_gender'])['trip_duration_min'].mean().reset_index()
    
# Plot using FacetGrid
g = sns.FacetGrid(avg_duration_gender, col='member_gender', height=5, aspect=1.2, sharey=True)
g.map_dataframe(sns.barplot, x='day_of_week', y='trip_duration_min', hue='bike_share_for_all_trip', palette='pastel')

# Add labels
for ax in g.axes.flat:
    for container in ax.containers:
        ax.bar_label(container, fmt='%.1f', label_type='edge', padding=3, fontsize=9)
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

g.add_legend(title='Bike Share for All')
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Day of Week", "Avg. Trip Duration (minutes)")
plt.tight_layout()
plt.show()
````

### 1. Why did you pick the specific chart?

Grouped bar chart - It compares multiple categories in a clear, side-by-side way.Also, it highlights differences in average trip duration based on whether users are part of the Bike Share for All program.

### 2. What is/are the insight(s) found from the chart?

Both Females and Males tend to take longer trips on Saturdays and Sundays compared to weekdays.Users Not enrolled in the "Bike Share for All" program (blue bars) consistently take longer trips.Females have higher average trip durations than Males during weekends.
